```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

Data = pd.read_csv('retailMarketing.csv')

labels = Data['Age']
data = Data.drop(['Age'],axis=1)

columns = data.columns
data = data.drop([columns[0]],axis=1)
data.head()
```

```python
columns = data.columns
cols_to_norm = [columns[0], columns[1], columns[2]]
data[cols_to_norm] = data[cols_to_norm].apply(lambda x: (x-x.min())/(x.max()-x.min()))
data.head()
```

```python
ages = labels.unique()
target = np.zeros(len(labels))
for i in range(len(target)):
    if labels[i]==ages[0]:
        target[i]=0
    elif labels[i]==ages[1]:
        target[i]=1
    elif labels[i]==ages[2]:
        target[i]=2
        
Target = pd.DataFrame(target)
Target.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data, Target, test_size=0.3, random_state=42, stratify=Target)
```

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix, classification_report
```

```python
svc = svm.SVC()
svc.fit(X_train, Y_train.values.ravel())
Y_pred = svc.predict(X_test)
print(confusion_matrix(Y_test, Y_pred, [0, 1, 2]))
print(classification_report(Y_test, Y_pred))
```

```python
Lisvm = svm.LinearSVC(C=0.5)
Lisvm.fit(X_train, Y_train.values.ravel())
Y_pred2 = Lisvm.predict(X_test)
print(confusion_matrix(Y_test, Y_pred2, [0, 1, 2]))
print(classification_report(Y_test, Y_pred2))
```

```python
Lisvm = svm.LinearSVC(C=2)
Lisvm.fit(X_train, Y_train.values.ravel())
Y_pred2 = Lisvm.predict(X_test)
print(confusion_matrix(Y_test, Y_pred2, [0, 1, 2]))
print(classification_report(Y_test, Y_pred2))
```
